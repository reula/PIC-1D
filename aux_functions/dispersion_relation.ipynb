## Dispersion Relation for Landau damping.

```julia
using SpecialFunctions, Plots
```

$$
f(k,\omega) = k^2 + 1+  i*\sqrt{\pi/2}[erf(\frac{i\omega}{k\sqrt{2}}) + 1]\frac{\omega}{k}e^{-(\omega/k)^2/2}
$$

```julia
dis(k,o)= k^2+1 + im*sqrt(pi/2)*(erf(im*o/k/sqrt(2)) + 1)*o/k*exp(-(o/k)^2/2)
```

```julia
dis(0.35,1.22-im*0.034)
#dis(0.01,1.00015)
```

To apply Newton's method we need the derivative of dis with respect to $\omega$

```julia
ddis(k,o) = -o/k^2 +
im*sqrt(pi/2)*(erf(im*o/k/sqrt(2))+1)*exp(-(o/k)^2/2)*(1-(o/k)^2)/k 
```

```julia
op = 1:0.01:2
plot(op,real.(dis.(0.35,op .- 0.034318050858299844im)),label="real dis")
plot!(op,imag.(dis.(0.35,op .- 0.034318050858299844im)),label="imag dis")
plot!(op,real.(ddis.(0.35,op .- 0.034318050858299844im)),label="real ddis")
plot!(op,imag.(ddis.(0.35,op .- 0.034318050858299844im)),label="imag dis")
```

```julia
N = 10
k = 0.5
o = 1.0
for i in 1:N
    o = o - dis(k,o)/ddis(k,o)
    println(o)
end
```
